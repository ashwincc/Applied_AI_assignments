TASK 1

```python
clf = SGDClassifier(loss='log_loss')
```

```python
clf.fit(X, Y)
```

```python
coeff_dict = {}
for coeff, feat in zip(clf.coef_[0,:],['f1','f2','f3']): # # PRINTING THE WEIGHT COEFFICENT SVM
    coeff_dict[feat] = coeff
print(coeff_dict)
```

```python
svm_clf = SGDClassifier(loss='hinge')
```

```python
svm_clf.fit(X,Y)
```

```python
coeff_dict = {}
for coeff, feat in zip(svm_clf.coef_[0,:],['f1','f2','f3']): # # PRINTING THE WEIGHT COEFFICENT SVM
    coeff_dict[feat] = coeff
print(coeff_dict)
```


Observations

    According to above results feature importance is f2>f1>f3 it is proportional to above variance we calculated( var(F2)>>var(F1)>>Var(F3))
    Both in log classifier and SVM high variance is important feature.
